### Increase intensity of lena.png by 45 in grayscale

```python
img1 = cv2.imread('lena.png', cv2.IMREAD_GRAYSCALE)
img1 += 45
plt.imshow(img1, cmap = "gray")
plt.show()
```

### Corrected method for linear intensity increase (What might cause an error on direct linear increase of intensities?)

```python
img1 = cv2.imread('lena.png', cv2.IMREAD_GRAYSCALE)
cv2.add(img1,45)
plt.imshow(img1, cmap = "gray")
plt.show()
```

### Construct a frequency histogram for bell.jpg, and plot the histogram using plt.bar (search on google for syntax)

```python
img = cv2.imread('bell.jpeg',0)
  
histr = cv2.calcHist([img],[0],None,[256],[0,256])
  
plt.plot(histr)
plt.show()
```

### Perform thresholding on bell.jpg using the above constructed histogram. 

```python
img = cv2.imread('bell.jpeg',0)
ret, thresh1 = cv2.threshold(img, 115, 255, cv2.THRESH_BINARY)
plt.imshow(img, cmap = "gray")
plt.show()
plt.imshow(thresh1, cmap = "gray")
plt.show()
```

### Do a simple thresholding on the image sudoku.jpg, and find an apt threshold for the same using the above method. 

```python
img = cv2.imread('sudoku.jpg',0)

histr = cv2.calcHist([img],[0],None,[256],[0,256])
  
plt.plot(histr)
plt.show()

ret, thresh1 = cv2.threshold(img, 50, 255, cv2.THRESH_BINARY)
plt.imshow(thresh1, cmap = "gray")
plt.show()
```

### Plot the RGB histograms for the image lotus.jpg using inbuilt functions. (Look up cv2.calcHist)



```python
img = cv2.imread('lotus.jpg', cv2.IMREAD_UNCHANGED)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
color = ('r','g','b')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

### Perform thresholding on fingerprint.jpg and find the threshold. 



```python
img = cv2.imread('fingerprint.jpg',0)

histr = cv2.calcHist([img],[0],None,[256],[0,256])
  
plt.plot(histr)
plt.show()

ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
plt.imshow(thresh1, cmap = "gray")
plt.show()
```
